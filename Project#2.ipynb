STEP 3:
<ul>

<li>We import request and json</li>
<li>We create a function like get_stream shown below to stream in the news data from the API</li>
<li>In the get_stream(url) we :<ul>
    <li>Send the query URL to the API using request.get and set stream to True so that we can stream and sve it as r</li>
    <li>From the output of the above we iterrate over the lines</li>
    <li>We decode the line and use jsaon loads so that we get the data in a dictionary</li>
    <li>Now the dictionary contains some keys which differ according to the API we are using like status,number of articles etc.We need to see for the dictionary that has the the articles as a list of dictionaries</li>
    <li>In the list of dictionaries of aritcles there are many keys like source ,URL,Image URL,author,language,country,topic,content etc.We need to choose as our need.We choose the tilte to stream to our database.So for every dictionary in the list of articles we save the content in the title key.</li>
    <li>We check whether that title is already in the table by checking if the title is in the records list as said in STEP 2</li>
    <li>If the record is not in the said list .we execute the SQL command  mycursor.execute(query,(val,)) .Here query is as discussed above and then pass (val,) as the record we wanrt to insert in the database.And lastly perform mydb.commit() to save the changes to the database table </li>
    </ul>
</ul>
 Detailed procedure shown below   

```python
import requests
import json

query="INSERT INTO headline (title) VALUES (%s)"
def get_stream(url):
    r = requests.get(url, stream=True)
    for line in r.iter_lines(chunk_size=10):
        if line:
            decoded_line = line.decode('utf-8')
            dat=json.loads(decoded_line)['articles']
        for lines in dat:
            mycursor = mydb.cursor()
            val=lines['title']
            if (val,) not in records:
                mycursor.execute(query,(val,))
                mydb.commit()
url = ('http://newsapi.org/v2/everything?''q=India&''from=2020-08-01&''to=2019-08-01&''sortBy=popularity&''pageSize=100&''apiKey=6ad7a293f5534b97951d0377356ae7dd')
get_stream(url)
```
